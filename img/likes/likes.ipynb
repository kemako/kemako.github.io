```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15, 15), dpi=100)
# 円グラフを描画
x = np.array([10, 7, 6, 5, 3,3])
label = ["Robot","Cocking", "IoT", "Design", "Computer Vision","Web"]
colors = ['cornflowerblue', 'coral', 'lightseagreen', 'hotpink', 'sandybrown', 'indianred']
patches, texts = plt.pie(x, labels=label, counterclock=False, startangle=90, colors=colors, labeldistance=0.6, textprops={'color': "whitesmoke", 'weight': "bold"})
centre_circle = plt.Circle((0,0),0.3, fc="#f3f3f3",linewidth=1.25)
for t in texts:
    t.set_horizontalalignment('center')
    t.set_size(32)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.savefig("likes_fig.png", transparent=True)
plt.show()
```
